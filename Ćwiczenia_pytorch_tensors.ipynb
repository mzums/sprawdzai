# Dwuwymiarowe tensory
W przypadku Pythonowych list, możliwe jest umieszczenie list w środku list np. tak jak poniżej: <br>
[[1,2,3],[2,3,4],[1]] <br>
Jak można zauważyć, pierwszym elementem powyższej listy jest lista [1,2,3], drugim jest lista [2,3,4] i trzecim jest lista [1]. Tensory również mogą zawierać inne struktury podobnie jak listy. Co więcej, okazuje się, że jeśli chcecie w jakikolwiek sposób wyciągnąć dane potrzebne do trenowania modeli, bardzo często będziecie musieli na takich dwuwymiarowych tensorach operować. Tensory różnią się jednak pod pewnym względem od Pythonowych list. W Pythonie, jeśli lista zawiera inne listy, każda z list, które ona zawiera mogą być różnej długości, jak na powyższym przykładzie, gdzie pierwsza lista ma długość 3, druga długość 3 i trzecia długość 1. W kontekście tensorów, wszystkie wewnętrzne struktury muszą być tej samej długości.

# Łączenie dwuwymiarowych tensorów
Funkcja torch.cat((tensor1, tensor2), dim=a) potrafi nie tylko łączyć ze sobą dwa jednowymiarowe tensory, ale również dwuwymiarowe tensory. Parametr dim odnosi się do konkretnego wymiaru, w którym dwa tensory są ze sobą łączone. Jest to może trochę trudny koncept do zrozumienia (mimo wszystko jak to możliwe, że można coś połączyć na kilka sposobów???), ale mam nadzieję, że wszystko rozjaśni się w praniu.

W PyTorch dim=0 oznacza pierwszy wymiar (wiersze), a dim=1 oznacza drugi wymiar (kolumny).

## Zadanie 6
Funkcja torch.randn((długość1, długość2)) zwraca dwuwymiarowy tensor o wymiarach długość1 x długość2, zawierający losowe wartości z dystrybucji normalnej. Sposób w jaki możesz rozumieć te losowe wartości to, że są one dowolnymi wartościami w zakresie od -nieskończoność do nieskończoność takimi, że prawdopodobieństwo, że konkretna liczba będzie wylosowana jest tym większe, im bliższa jest ona 0. Np. wartość 1 jest bardziej prawdopodobna niż wartość 2, natomiast wartość -2 jest bardziej prawdopodobna niż wartość -3. Utwórz dwa tensory o nazwach array1 i array2, oba o wymiarach 2 x 6, a następnie połącz je ze sobą na dwa możliwe sposoby, tworząc array3 i array4. Następnie, wypisz array3 i array4 oraz ich kształty. Kształty wypiszesz poprzez napisanie print(array3.shape) oraz print(array4.shape). Array3 powinno mieć kształt 4 x 6, natomiast array4 powinno mieć kształt 2 x 12.

```python
array1 = torch.randn((2,6))
array2 = torch.randn((2,6))
```

```python
array3 = torch.cat((array1, array2), dim=0)
print(array3)
print(array3.shape)
```

```python
array4 = torch.cat((array1, array2), dim=1)
print(array4)
print(array4.shape)
```

Wyjściem programu powinny być dwa tensory: pierwszy o wymiarach (4, 6), a drugi o wymiarach (2, 12) wraz z ich kształtami.

# Trójwymiarowe tensory
Dwuwymiarowe tensory można interpretować jako lista zawierająca listy. Trójwymiarowe tensory interpretuje się jako lista zawierająca listy, które same zawierają listy.

Jak zobaczysz w poniższych zadaniach, trójwymiarowe tensory działają analogicznie do dwuwymiarowych.
