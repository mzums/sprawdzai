## Zakładka zgłoszeń

Każde zgłoszenie do tego zadania możesz wysłać na jeden z dwóch zbiorów - walidacyjny lub testowy.

**Zbiór walidacyjny** zawiera te same dane co `snails_valid.pt` oraz `labels_valid.pt`, czyli jest tylko sposobem *walidacji*, czy Twoje rozwiązanie działa poprawnie na naszym systemie (ten notatnik i tak wypisze wynik na zbiorze walidacyjnym). Dlatego nie ma żadnych limitów na wysyłanie zgłoszeń na zbiór walidacyjny (poza tymi, które obowiązaują na całej platformie).

Z kolei **zbiór testowy** jest inny. Nie masz dostępu do tych danych - są one wyłącznie na naszych serwerach. Dzięki temu możemy dobrze sprawdzić, czy Twoje rozwiązanie faktycznie generalizuje na nowych danych. To wynik na zbiorze testowym jest faktycznym wynikiem za to zadanie. Na wysyłanie zgłoszeń na zbiór testowy obowiązuje dodatkowy limit pięciu zgłoszeń dziennie (w zależności od zadania, te limity mogą oczywiście być też inne) - dzięki temu masz pewność, że nie przetrenowałeś swojego algorytmu tak, aby działał specyficznie pod zbiór testowy (co byłoby możliwe po wysłaniu wystarczającej liczby zgłoszeń!).

Po wysłaniu, Twoje zgłoszenie pojawi się w zakładce **zgłoszenia** do tego zadania.
Możesz je tam **kliknąć**, aby zobaczyć jego szczegóły, lub je **anulować**.

Dla ułatwienia, możesz **przesłać** swoje zgłoszenie na zbiór testowy lub walidacyjny. Takie zgłoszenie na odpowiadającym zbiorze wyświetli się wtedy na lewo lub prawo od oryginalnego. Automatycznie wyśle ono te same pliki które były wysłane w oryginalnym zgłoszeniu, ale uruchomi je na drugim zbiorze. Dzięki temu możesz wysłać zgłoszenie na zbiór walidacyjny aby sprawdzić jego poprawność, a następnie ponownie przesłać je na zbiór testowy żeby zobaczyć faktyczny wynik i łatwo porównywać wyniki między tymi zbiorami.

W szczegółach zgłoszenia znajdziesz pola `STDOUT uruchomienia notatnika` oraz `STDERR uruchomienia notatnika`.
```python
X = torch.tensor(snails)

y = torch.tensor(labels, dtype=torch.float32)
y = y.unsqueeze(1)

model = nn.Sequential(
    nn.Linear(4, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1)
)

criterion = nn.BCEWithLogitsLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(3000):
    logits = model(X)
    loss = criterion(logits, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 300 == 0:
        print(epoch, loss.item())
```

```python
def predict(data):
    X_valid = torch.tensor(data, dtype=torch.float32)

    y_valid = torch.tensor(labels_valid, dtype=torch.float32)
    y_valid = y_valid.unsqueeze(1)

    with torch.no_grad():

        logits = model(X_valid)

        probs = torch.sigmoid(logits)

        preds = (probs > 0.5).float()
        preds = preds.numpy().flatten().astype(np.int64)

    return preds
```

# Kod oceniający
Ten kod ma na celu ocenienie twojego rozwiązania na dostępnym dla ciebie zbiorze walidacyjnym. Aby nie wliczał się on do limitu czasu, uruchomi się on jedynie u ciebie.

```python
def points(predictions, labels):
    #print(predictions[:20].numpy().flatten())
    predictions = predictions.astype(np.int64)
    labels = labels.astype(np.int64)

    # dokładność
    accuracy = np.mean(predictions == labels)

    acc_min = 0.50
    acc_max = 0.95

    if accuracy <= acc_min:
        return 0, accuracy
    elif accuracy >= acc_max:
        return 100, accuracy
    else:
        score = (accuracy - acc_min) / (acc_max - acc_min) * 100
        return int(score), accuracy


if not FINAL_EVALUATION_MODE:
    pred = predict(snails_valid)

    wynik, dokladnosc = points(pred, labels_valid)

    # Zwyczajne wykonania funkcji 'print' zapisywane są w strumieniu STDOUT, jest on widoczny po wysłaniu notatnika
    # Błędy, wywołania 'print' z argumentem file=sys.stderr oraz paski postępu tqdm zapisywane są w strumieniu STDERR, jest on również widoczny po wysłaniu notatnika
    print(f"Twój wynik na zbiorze walidacyjnym: {wynik} pkt")
    print(f"Dokładność: {dokladnosc * 100:.2f}%")
    
```
